```python
nfit = 789
nval = 9
pred = 39

tas = np.load('../submit/data/saved/t2m_obs.npy', mmap_mode = 'r')[(nfit + pred),]
pr = np.load('../submit/data/saved/pr_obs.npz.npy', mmap_mode = 'r')[(nfit + pred),]

tas_nngp = np.load('./submit/pred/nngp_pred.npy', mmap_mode = 'r')[pred]
tas_lm = np.load('./submit/pred/lm_pred.npy', mmap_mode = 'r')[(nval+pred)]
tas_ens = np.load('./submit/pred/ens_pred.npy', mmap_mode = 'r')[(nval+pred)]
```

```python
fig = plt.figure(constrained_layout = True, figsize = (10, 6))
gs = GridSpec(3, 4, width_ratios = [1, 1, 1, 0.05], figure = fig)

titles = ['ERA5 Reanalysis (Jan. 2015)', 'NN-GPR Prediction', 'Ens. Average (EA)']
coords = [[150, 940], [390, 1105], [170, 280]]
loc = ['Western \nUnited States', 'Andes\nMountains', 'Tibetan Plateau']

for k in trange(3, leave = False):
    y1 = coords[k][0]
    y2 = y1 + 100
    x1 = coords[k][1]
    x2 = x1 + 140
    
    data = [tas[y1:y2,x1:x2], tas_nngp[y1:y2,x1:x2],tas_ens[y1:y2,x1:x2]]
    
    ax = plt.subplot(gs[k, 0])
    ax.imshow(data[0], cmap = 'RdBu_r', aspect = 'auto')
    ax.set_ylabel(loc[k], fontsize = 15)
    ax.set_yticks([])
    ax.set_xticks([])
    
    if k == 0:
        ax.set_title(titles[0], fontsize = 15)
    
    for p in range(1, 3):
        ax = plt.subplot(gs[k, p])
        im = ax.imshow(data[p], cmap = 'RdBu_r', aspect = 'auto')
        ax.set_yticks([])
        ax.set_xticks([])
        if k == 0:
            ax.set_title(titles[p], fontsize = 15)
        
    ax = plt.subplot(gs[k, 3])
    cbar = plt.colorbar(im, cax=ax)
    cbar.ax.tick_params(labelsize=10)
```
