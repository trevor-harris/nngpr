```python
%%capture

# spatial region to extract
br = 50
tr = br + 160
lc = 161
rc = lc + 240

rcp_files = sorted(os.listdir('../data/RCM/'))
rcp_offset = 81-72

# t2m
na_rcp = []
for f in rcp_files:
    if f[0:4] == 'temp':
        ncdata = nc.Dataset('../data/RCM/' + f, "r")
        ncdata = ncdata.variables['temp'][rcp_offset:,br:tr,lc:rc]
        ncdata = np.array(np.flip(np.flip(ncdata, axis = (1, 2)), 2))
        na_rcp.append(ncdata)
        
tas_rcp = np.array(na_rcp)[0:2,0:pred_int]
tas_rcp = np.mean(tas_rcp, axis = 0)
# tas_rcp = np.mean(tas_rcp, axis = 0) + 1.5


# pr
na_rcp = []
for f in rcp_files:
    if f[0:4] == 'prec':
        ncdata = nc.Dataset('../data/RCM/' + f, "r")
        ncdata = ncdata.variables['prec'][rcp_offset:,br:tr,lc:rc]
        ncdata = np.array(np.flip(np.flip(ncdata, axis = (1, 2)), 2))
        na_rcp.append(ncdata)
        
pr_rcp = np.array(na_rcp)[0:2,0:pred_int]
pr_rcp = np.mean(pr_rcp, axis = 0)
# pr_rcp = np.mean(pr_rcp, axis = 0) - 0.25

tas_nngp = np.load('./tas_predictions/nngp_pred.npy', mmap_mode = 'r')[0:pred_int,lat1:lat2,lon1:lon2]
tas_lm = np.load('./tas_predictions/lm_pred.npy', mmap_mode = 'r')[nval:(nval+pred_int),lat1:lat2,lon1:lon2]
pr_nngp = np.load('./pr_predictions/nngp_pred.npy', mmap_mode = 'r')[0:pred_int,lat1:lat2,lon1:lon2]
pr_lm = np.load('./pr_predictions/lm_pred.npy', mmap_mode = 'r')[nval:(nval+pred_int),lat1:lat2,lon1:lon2]
```

```python
## t2m
tas_mse = [np.mean((tas_nngp - tas)**2 * scale, axis = (1, 2)),
              np.mean((tas_lm - tas)**2 * scale, axis = (1, 2)),
              np.mean((tas_rcp - tas)**2 * scale, axis = (1, 2))]

tas_bias = [np.mean((tas_nngp - tas) * scale, axis = (1, 2)),
              np.mean((tas_lm - tas) * scale, axis = (1, 2)),
              np.mean((tas_rcp - tas) * scale, axis = (1, 2))]

tas_max = [np.max(np.abs(tas_nngp - tas) * scale, axis = (1, 2)),
              np.max(np.abs(tas_lm - tas) * scale, axis = (1, 2)),
              np.max(np.abs(tas_rcp - tas) * scale, axis = (1, 2))]

tas_ssim = [np.array([ssim(tas[i].astype('float32'), tas_nngp[i].astype('float32')) for i in range(pred_int)]),
               np.array([ssim(tas[i].astype('float32'), tas_lm[i].astype('float32')) for i in range(pred_int)]),
               np.array([ssim(tas[i].astype('float32'), tas_rcp[i].astype('float32')) for i in range(pred_int)])]

## pr
pr_mse = [np.mean((pr_nngp - pr)**2 * scale, axis = (1, 2)),
              np.mean((pr_lm - pr)**2 * scale, axis = (1, 2)),
              np.mean((pr_rcp - pr)**2 * scale, axis = (1, 2))]

pr_bias = [np.mean((pr_nngp - pr) * scale, axis = (1, 2)),
              np.mean((pr_lm - pr) * scale, axis = (1, 2)),
              np.mean((pr_rcp - pr) * scale, axis = (1, 2))]

pr_max = [np.max(np.abs(pr_nngp - pr) * scale, axis = (1, 2)),
              np.max(np.abs(pr_lm - pr) * scale, axis = (1, 2)),
              np.max(np.abs(pr_rcp - pr) * scale, axis = (1, 2))]

pr_ssim = [np.array([ssim(pr[i].astype('float32'), pr_nngp[i].astype('float32')) for i in range(pred_int)]),
               np.array([ssim(pr[i].astype('float32'), pr_lm[i].astype('float32')) for i in range(pred_int)]),
               np.array([ssim(pr[i].astype('float32'), pr_rcp[i].astype('float32')) for i in range(pred_int)])]
```

```python
def signed_maxdiff(x, y):
    m1 = np.max((x - y) * scale, axis = (1, 2))
    m2 = np.min((x - y) * scale, axis = (1, 2))
    
    md = []
    for i, j in zip(m1, m2):
        md.append(i if i > np.abs(j) else j)
        
    return np.array(md)
```

```python
tas_metrics = [tas_mse, tas_ssim, tas_bias, tas_max]
pr_metrics = [pr_mse, pr_ssim, pr_bias, pr_max]
```

```python
tas_msemaps = [np.mean((tas_nngp - tas)**2, axis = 0),
              np.mean((tas_lm  - tas)**2, axis = 0),
              np.mean((tas_rcp  - tas)**2, axis = 0)]

pr_msemaps = [np.mean((pr_nngp - pr)**2, axis = 0),
              np.mean((pr_lm  - pr)**2, axis = 0),
              np.mean((pr_rcp  - pr)**2, axis = 0)]
```

```python
lat1 = 102
lat2 = lat1 + 160
lon1 = 914
lon2 = lon1 + 240

nfit = 789
nval = 9
pred_int = 70

scale = np.cos(np.linspace(math.pi/2, -math.pi/2, 721))[lat1:lat2]
scale /= np.mean(scale)
scale = scale.reshape((1, -1, 1))

tas = np.load('../data/saved/t2m_obs.npy', mmap_mode = 'r')[0:nfit,lat1:lat2,lon1:lon2]
tas_mean = np.mean(tas * scale)

pr = np.load('../data/saved/pr_obs.npz.npy', mmap_mode = 'r')[0:nfit,lat1:lat2,lon1:lon2]
pr_mean = np.mean(pr * scale)
```

```python
%%capture

# spatial region to extract
br = 50
tr = br + 160
lc = 161
rc = lc + 240

rcp_files = sorted(os.listdir('../data/RCM/'))
rcp_offset = 81-72

# t2m
na_rcp = []
for f in rcp_files:
    if f[0:4] == 'temp':
        ncdata = nc.Dataset('../data/RCM/' + f, "r")
        ncdata = ncdata.variables['temp'][rcp_offset:,br:tr,lc:rc]
        ncdata = np.array(np.flip(np.flip(ncdata, axis = (1, 2)), 2))
        na_rcp.append(ncdata)
        
na_rcp = np.array(na_rcp)[0:2,:] - tas_mean
tas_rcp = np.mean(na_rcp, axis = 0)


# pr
na_rcp = []
for f in rcp_files:
    if f[0:4] == 'prec':
        ncdata = nc.Dataset('../data/RCM/' + f, "r")
        ncdata = ncdata.variables['prec'][rcp_offset:,br:tr,lc:rc]
        ncdata = np.array(np.flip(np.flip(ncdata, axis = (1, 2)), 2))
        na_rcp.append(ncdata)
        
pr_rcp = np.array(na_rcp)[0:2,:] - pr_mean
pr_rcp = np.mean(pr_rcp, axis = 0)

tas_nngp = np.load('./tas_predictions/nngp_pred.npy', mmap_mode = 'r')[:,lat1:lat2,lon1:lon2] - tas_mean
tas_lm = np.load('./tas_predictions/lm_pred.npy', mmap_mode = 'r')[nval:,lat1:lat2,lon1:lon2] - tas_mean
tas_ens = np.load('./tas_predictions/ens_pred.npy', mmap_mode = 'r')[nval:,lat1:lat2,lon1:lon2]  - tas_mean
 
pr_nngp = np.load('./pr_predictions/nngp_pred.npy', mmap_mode = 'r')[0:,lat1:lat2,lon1:lon2] - pr_mean
pr_lm = np.load('./pr_predictions/lm_pred.npy', mmap_mode = 'r')[nval:,lat1:lat2,lon1:lon2] - pr_mean
pr_ens = np.load('./pr_predictions/ens_pred.npy', mmap_mode = 'r')[nval:,lat1:lat2,lon1:lon2] - pr_mean
```

```python
def yearly_mean(x):
    return np.array([np.mean(x[(i*12):((i+1)*12)] * scale) for i in range(85)])

def climate_mean(x, i):
    return np.mean(x[(i*20):((i+1)*20)] * scale, axis = 0)
```

```python
fig = plt.figure(constrained_layout = True, figsize = (18, 11))
gs = GridSpec(3, 5, width_ratios = [1, 1, 1, 1, 0.05], figure = fig)

data_crs = ccrs.PlateCarree(central_longitude=180)
plot_crs = ccrs.Robinson(central_longitude=180)
gcm_extent = (48.25, 108.25, 24.5, 64.5)

s = np.linspace(2015, 2100, 85)

ax = plt.subplot(gs[0,0:2])
ax.plot(s, yearly_mean(tas_ens), c = 'grey', label = 'EA')
ax.plot(s, yearly_mean(tas_lm), label = 'LM')
ax.plot(s, yearly_mean(tas_rcp), c = 'black', label = 'RCM')
ax.plot(s, yearly_mean(tas_nngp), label = 'NN-GPR')
ax.set_title(f'T2M anomalies (\u2103)', fontsize = 23)
ax.tick_params(axis='both', which='major', labelsize = 20)
ax.legend(fontsize = 12)

ax = plt.subplot(gs[0,2:4])
ax.plot(s, yearly_mean(pr_ens), c = 'grey', label = 'EA')
ax.plot(s, yearly_mean(pr_lm), label = 'LM')
ax.plot(s, yearly_mean(pr_rcp), c = 'black', label = 'RCM')
ax.plot(s, yearly_mean(pr_nngp), label = 'NN-GPR')
ax.set_title(f'PR anomalies (mm)', fontsize = 23)
ax.tick_params(axis='both', which='major', labelsize = 20)
# ax.set_ylabel('PR anomaly (mm/day)', fontsize = 20)
ax.legend(fontsize = 12)

divnorm = colors.TwoSlopeNorm(vmin=-10, vcenter=0, vmax=10)
cmap = plt.get_cmap('RdBu_r', 9)
for i in trange(4, leave = False):
    ax = plt.subplot(gs[1, i], projection=data_crs)
    im = ax.imshow(climate_mean(tas_nngp, i) - climate_mean(tas_rcp, i),
                   cmap = cmap,
                   norm = divnorm,
                   extent = gcm_extent,
                   transform = data_crs)
    ax.coastlines()
    ax.set_aspect('auto', adjustable=None)
    ax.set_title(f'NN-GPR - RCM\n{2020 + 10*(2*i)} - {2020 + 10*(2*i+2)} (T2M)', fontsize = 17)
#     ax.set_title(f'T2M - temporal mean ({2020 + 10*(2*i)} - {2020 + 10*(2*i+2)})', fontsize = 12)

ax = plt.subplot(gs[1, 4])
cbar = plt.colorbar(im, cax=ax)
cbar.ax.tick_params(labelsize=12)
cbar.ax.set_title('\u2103', fontsize = 15, pad = 10)
    
divnorm = colors.TwoSlopeNorm(vmin=-5, vcenter=0, vmax=5)
cmap = plt.get_cmap('BrBG', 9)
for i in range(4):
    ax = plt.subplot(gs[2, i], projection=data_crs)
    im = ax.imshow(climate_mean(pr_nngp, i) - climate_mean(pr_rcp, i),
                   cmap = cmap,
                   norm = divnorm,
                   extent = gcm_extent,
                   transform = data_crs)
    ax.coastlines()
    ax.set_aspect('auto', adjustable=None)
    ax.set_title(f'NN-GPR - RCM\n{2020 + 10*(2*i)} - {2020 + 10*(2*i+2)} (PR)', fontsize = 17)
#     ax.set_title(f'PR - temporal mean ({2020 + 10*(2*i)} - {2020 + 10*(2*i+2)})', fontsize = 12)
    
ax = plt.subplot(gs[2, 4])
cbar = plt.colorbar(im, cax=ax)
cbar.ax.tick_params(labelsize=12)
cbar.ax.set_title('mm', fontsize = 15, pad = 10)

ax = plt.subplot(gs[0, 0:2])
ax.text(0.035, 0.94, 'a', fontsize=25, fontweight='bold', transform=plt.gcf().transFigure)

ax = plt.subplot(gs[0, 2:4])
ax.text(0.49, 0.94, 'b', fontsize=25, fontweight='bold', transform=plt.gcf().transFigure)

ax = plt.subplot(gs[1, 0])
ax.text(0.035, 0.59, 'c', fontsize=25, fontweight='bold', transform=plt.gcf().transFigure)

ax = plt.subplot(gs[2, 0])
ax.text(0.035, 0.27, 'd', fontsize=25, fontweight='bold', transform=plt.gcf().transFigure)

plt.show()
```
